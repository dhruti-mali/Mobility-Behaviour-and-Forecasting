```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectKBest, f_regression

df = data.copy()
# Drop the columns you specified
cols_to_drop = ['legid', 'userid', 'tripid', 'os', 'os_version', 'model']
df = df.drop(columns=cols_to_drop)

# Take a sample of your data if it's large
sample_fraction = 0.1
df_sample = df.sample(frac=sample_fraction, random_state=42)

# Separate the features and target variable
X = df_sample.drop('mood_rating', axis=1)
y = df_sample['mood_rating']

# Impute missing values with mean
# One-hot encoding categorical features
X = pd.get_dummies(X, drop_first=True)
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Train a Random Forest model
clf = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
clf.fit(X_imputed, y)

# Get feature importances and print them
feature_importances = clf.feature_importances_
important_features = pd.Series(feature_importances, index=X.columns).sort_values(ascending=False)
print(important_features)
```
